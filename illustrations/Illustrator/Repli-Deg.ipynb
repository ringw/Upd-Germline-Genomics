```r
library(cowplot)
library(dplyr)
library(ggplot2)
library(scales)
library(Seurat)
library(targets)
library(viridis)
```

```r
Upd_sc <- readRDS("../_targets/objects/Upd_sc")
Upd_decontX <- readRDS("../_targets/objects/Upd_decontX")
shuffle_feature_plot <- readRDS("../_targets/objects/shuffle_feature_plot")
Upd_sc[["DECONTX"]] <- Upd_decontX
```

```r
Upd_sc_feature_plot <- function(Upd_sc, gene, cells, assay = "RNA") {
  DefaultAssay(Upd_sc) <- assay
  dplyr_rename_lookup_gene = setNames("LogNormalize", gene)
  gene.data = FetchData(Upd_sc, c("umap_1", "umap_2", gene, "ident")) %>%
    plyr::rename(all_of(dplyr_rename_lookup_gene))
  gene.data <- gene.data[cells, ]
  gene.max.intensity <- NA
  gene.data %>% ggplot(
    aes(umap_1, umap_2, color=LogNormalize)
  ) + geom_point(
    size = 0.6, stroke = NA
  ) + scale_color_viridis_c(
    "Log Normalized",
    begin = 0.2,
    limits = c(0, gene.max.intensity), oob = squish,
    guide = guide_colorbar(barheight = 0.75, barwidth = 10)
  ) + scale_x_continuous(
    breaks=c(-10,0,10)
  ) + scale_y_continuous(breaks=c(-10,0,5)) + theme_cowplot(
    font_size = 28,
    line_size = 0.5
  ) + labs(
    tag=gene
  ) + theme(
    axis.title.x = element_text(margin = margin(0, 5.5, -20, 5.5)),
    axis.title.y = element_text(margin = margin(5.5, -20, 5.5, 5.5)),
    legend.title = element_text(margin = margin(-15, 5.5, 15, 5.5)),
    legend.box.margin = margin(15, 5.5, 5.5, 5.5),
    plot.margin = margin(50, 5.5, 5.5, 5.5),
    plot.tag = element_text(size = rel(1.37)),
    plot.tag.position = c(0.13, 1.08),
    aspect.ratio = 0.75,
    legend.position = "bottom"
  )
}
pdfFonts(sans = pdfFonts()$Helvetica)
postscriptFonts(sans = postscriptFonts()$Helvetica)
ggsave(
  "RNAseq-AGO3.pdf",
  Upd_sc_feature_plot(Upd_sc, "AGO3", shuffle_feature_plot, assay="DECONTX"),
  w=8,
  h=7.4,
  device=cairo_pdf
)
ggsave(
  "RNAseq-alphaTub84B.pdf",
  Upd_sc_feature_plot(Upd_sc, "alphaTub84B", shuffle_feature_plot, assay="DECONTX"),
  w=8,
  h=7.4,
  device=cairo_pdf
)
ggsave(
  "RNAseq-caps.pdf",
  Upd_sc_feature_plot(Upd_sc, "caps", shuffle_feature_plot, assay="DECONTX"),
  w=8,
  h=7.4,
  device=cairo_pdf
)
ggsave(
  "RNAseq-pb.pdf",
  Upd_sc_feature_plot(Upd_sc, "pb", shuffle_feature_plot, assay="DECONTX"),
  w=8,
  h=7.4,
  device=cairo_pdf
)
```
